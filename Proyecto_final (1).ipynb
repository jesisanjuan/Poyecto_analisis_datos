```python
Violencia_df['fec_hecho'] = pd.to_datetime(Violencia_df['fec_hecho'], format='%d/%m/%Y')
```

```python
Violencia_df['Ano'] = pd.to_datetime(Violencia_df['Ano'], format='%Y')
```

```python
Violencia_df = Violencia_df.drop(columns=['nom_upgd'])
```

```python
Violencia_df = Violencia_df.drop(columns=['MES'])
```

```python
Violencia_df.dtypes
```

```python
Violencia_df['Año'] = Violencia_df['fec_hecho'].dt.year
Violencia_df = Violencia_df[(Violencia_df['Año'] >= 2015) & (Violencia_df['Año'] <= 2023)]
denuncias_por_año = Violencia_df['Año'].value_counts().sort_index()
plt.figure(figsize=(14, 8)) #Tamaño de gráfica 15 ancho 10 largo
plt.plot(denuncias_por_año.index, denuncias_por_año.values, marker='o', linestyle='-', color='blue', label='Reportes de violencia')
# marker el punto acorde a la tendencia
plt.fill_between(denuncias_por_año.index, denuncias_por_año.values, color='slategrey', alpha=0.3)
#Diseños
plt.title('Tendencia de hechos de violencia intrafamiliar entre 2015 y 2023',fontsize=20) #Titulo
plt.xlabel('Años',fontsize=20) #Label x
plt.ylabel('Cantidad de denuncias',fontsize=20) #Label Y
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.legend() #Leyendas
plt.grid(True)
plt.tight_layout() #Orden
plt.show() #Mostrar
```

La gráfica presentada muestra la evolución de los casos de violencia intrafamiliar y de género desde 2015 hasta 2023.

**Tendencia General:**

**Violencia de Género:** La tendencia para la violencia de género también muestra un aumento progresivo. Es crucial considerar que el aumento en los reportes podría deberse a una mejor detección, un mayor acceso a servicios de denuncia y un cambio en la percepción social que anima a las víctimas a hablar.

```python
plt.figure(figsize=(12, 12))
frecuencia_municipio= Violencia_df['Municipio'].value_counts()
frecuencia_municipio=frecuencia_municipio.head(9)
ax = frecuencia_municipio.plot(kind='bar', color='green', edgecolor='black')
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                    ha='center', va='center', xytext=(0, 10), textcoords='offset points',fontsize=15)

plt.title('Municipio ocurrencia hechos de violencia intrafamiliar y de genero',fontsize=18)
plt.xlabel('Municipio',fontsize=18)
plt.ylabel('Hechos registrados',fontsize=18)
plt.xticks(fontsize=18)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 12))
frecuencia_area= Violencia_df['area'].value_counts()
frecuencia_area=frecuencia_area.head(3)
ax = frecuencia_area.plot(kind='bar', color='violet', edgecolor='black')
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                    ha='center', va='center', xytext=(0, 10), textcoords='offset points',fontsize=15)

plt.title('Frecuencia de hechos en cada zona de Santander',fontsize=18)
plt.xlabel('Area del departamento',fontsize=18)
plt.ylabel('Hechos registrados',fontsize=18)
plt.xticks(fontsize=18)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 12))
frecuencia_naturaleza= Violencia_df['naturaleza'].value_counts()
frecuencia_naturaleza=frecuencia_naturaleza.head(9)
ax = frecuencia_naturaleza.plot(kind='bar', color='grey', edgecolor='black') #diseño
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                    ha='center', va='center', xytext=(0, 10), textcoords='offset points',fontsize=15)

plt.title('Naturaleza de los hechos de violencia registrados',fontsize=18)
plt.xlabel('Naturaleza',fontsize=18)
plt.ylabel('Hechos registrados',fontsize=18)
plt.xticks(fontsize=18)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 12))
frecuencia_grupo_edad= Violencia_df['Grupo edad'].value_counts()
frecuencia_grupo_edad=frecuencia_grupo_edad.head(6)
ax = frecuencia_grupo_edad.plot(kind='bar', color='darkblue', edgecolor='black')
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                    ha='center', va='center', xytext=(0, 10), textcoords='offset points',fontsize=15)

plt.title('Frecuencia de hechos segun grupo de edad',fontsize=18)
plt.xlabel('Grupo de edad',fontsize=18)
plt.ylabel('Hechos registrados',fontsize=18)
plt.xticks(fontsize=18)

plt.tight_layout()
plt.show()
```
